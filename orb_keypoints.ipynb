```python
import numpy as np
import cv2
import urllib
from matplotlib import pyplot as plt
from skimage import io
url = "https://upload.wikimedia.org/wikipedia/en/thumb/7/7d/Lenna_%28test_image%29.png/220px-Lenna_%28test_image%29.png"
```

```python
print "downloading %s" % (url)
img = io.imread(url)
#cv2.imshow("Correct", cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
#cv2.waitKey(0)
```

```python
img.shape[0]
```

```python
h,w = img.shape[:2]
print(img.shape)
fig = plt.figure(figsize = (10,10))
iplot = fig.add_subplot(111)
iplot.imshow(img)
plt.show()
```

```python
cv2.__version__
```

```python
# Initiate STAR detector
orb = cv2.ORB()

# find the keypoints with ORB (detectors)
kp = orb.detect(img,None)

# compute the descriptors with ORB
kp, des = orb.compute(img, kp)
```

```python
# draw only keypoints location,not size and orientation
img2 = cv2.drawKeypoints(img, kp, None, color=(0,255,0), flags=cv2.DRAW_MATCHES_FLAGS_DEFAULT)
fig = plt.figure(figsize = (10,10))
iplot = fig.add_subplot(111)
iplot.imshow(img2),plt.show()
```
